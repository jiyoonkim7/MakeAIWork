```python
intervals = [45, 50, 55, 60, 65, 70, 75,80]
col = df['Work']
df['Time'] = pd.cut(x=col, bins=intervals)

# To be able to look at the result stored in the variable
df['Time'] 
```

```python
df.groupby('Time')['Time'].count()
```

```python
# The _oh means one-hot
df_oh = pd.get_dummies(df)
# Display the one-hot encoded dataframe
df_oh 
```

```python
df_oh = df_oh.drop(['Power'], axis=1)
df_oh.shape
```

```python
import seaborn as sns
# Seaborn is een data-visualisatie package

# Dropping CustomerID column from data 
df = df.drop('Index', axis=1)

sns.pairplot(df)
```

```python
sns.scatterplot(x=df['Money'],
                y=df['Health'])
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))
plt.title("Customers Dendrogram")

# Selecting Annual Income and Spending Scores by index
selected_data = df_oh.iloc[:, 1:3]
clusters = shc.linkage(selected_data, 
            method='ward', 
            metric="euclidean")
shc.dendrogram(Z=clusters)
plt.show()
```

```python
plt.figure(figsize=(10, 7))
plt.title("Gender_equality")
clusters = shc.linkage(selected_data, 
            method='ward', 
            metric="euclidean")
shc.dendrogram(clusters)
plt.axhline(y = 20, color = 'r', linestyle = '-')
```
