## Import

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as cp
```

## Predefine

```python
def plot_init():
    
    fig, ax = plt.subplots()
    
    return ax


def plot_add(ax, x, y):
    
    ax.plot(x, y, linewidth=2.0)


def plot_show():
    
    plt.show()


def plot_graph(x, y):
    
    ax = plot_init()
    plot_add(ax, x, y)
    plot_show()
```

## Lineair verband
### Model
huizenprijs = vastgoed offset + prijs per oppervlak x oppervlak

p = 25.000 + 5.000 x oppervlak

### Stap 1: Maak huizenprijs model

```python
# All in euro
offset = 25_000
price_per_m2 = 5_000

# Create m2
house_m2 = np.arange(50, 250)

# Add some noise
house_prices = offset + price_per_m2 * house_m2

# Add some noise
house_prices = np.random.normal(house_prices, 50_000)

# Just making sure
plot_graph(house_m2, house_prices)
```

### Stap 2: Bekijk interval [50, 75] m2

```python
# Consider interval [50, 75] m2
house_m2_50_75 = house_m2[0:25]
house_prices_50_75 = house_prices[0:25]

# Plot interval
plot_graph(house_m2_50_75, house_prices_50_75)
```

### Stap 3: Fit data

#### Use the curve fitting

https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html

```python
# Define our model
def model(x, a, b):
    
    return a + np.array(b * x)

# Use scipy optimize package
house_prices_fit, var = cp.optimize.curve_fit(model, house_m2_50_75, house_prices_50_75)

print(house_prices_fit)

# Create fit
house_prices_prediction = house_prices_fit[0] + house_prices_fit[1] * house_m2_50_75

# Plot fit
ax = plot_init()
plot_add(ax, house_m2_50_75, house_prices_50_75)
plot_add(ax, house_m2_50_75, house_prices_prediction)
plot_show()
```

### Stap 4: Extrapoleer

```python
# Create fit
house_prices_prediction = house_prices_fit[0] + house_prices_fit[1] * house_m2

# Plot prediction for whole range
ax = plot_init()
plot_add(ax, house_m2, house_prices)
plot_add(ax, house_m2, house_prices_prediction)
plot_show()
```

## Polynomiale verbanden

### Model



massa bolmannetje = dichtheid x bolvormigheid + benen + armen + hoofd

mass_bolman = 1.03 x radius^3 + 12.4 + 8.7 + 3.2

### Stap 1: Maak model bolmannetje

```python
# Density in kg per m3
density = 1.03

# Other in kg
mass_legs = 12.4
mass_arms = 8.7
mass_head = 3.2

# Create radii
bol_radii = np.arange(1, 10, 0.1)

# Add some noise
bolman_mass = 1.3 * (bol_radii**3) + mass_legs + mass_arms + mass_head

# Add some noise
bolman_mass = np.random.normal(bolman_mass, 20.3)

# Just making sure
plot_graph(bol_radii, bolman_mass)
```

### Stap 2: Fit data

```python
# Define our model
def model(x, a, b):
    
    return a + np.array(b * x**3)

# Use scipy optimize package
bolman_mass_fit, var = cp.optimize.curve_fit(model, bol_radii, bolman_mass)

print(bolman_mass_fit)

# Create fit
bolman_mass_prediction = bolman_mass_fit[0] + bolman_mass_fit[1] * bol_radii**3

# Plot fit
ax = plot_init()
plot_add(ax, bol_radii, bolman_mass)
plot_add(ax, bol_radii, bolman_mass_prediction)
plot_show()
```
